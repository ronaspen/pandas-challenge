```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# print the merged dataframe
school_data_complete
```

### Formatting Scores Function

```python
common_format_mappings = {
    "Average Maths Score": "{:.2f}",
    "Average Reading Score": "{:.2f}",
    "% Passing Maths": "{:.2f}%",
    "% Passing Reading": "{:.2f}%",
    "% Overall Passing": "{:.2f}%"
}

def format_mappings(df, mappings):
    """Format dataframe columns based on given mappings."""
    formatted_df = df.copy()
    for column, format_string in mappings.items():
        if column in df.columns:
            formatted_df[column] = df[column].map(format_string.format)
    return formatted_df
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# total number of schools
total_schools = school_data_complete["school_name"].nunique()

# total number of students
total_students = school_data_complete["student_name"].count()

# total budget
total_budget = school_data["budget"].sum()

# average maths score
average_maths_score = school_data_complete["maths_score"].mean()

# average reading score
average_reading_score = school_data_complete["reading_score"].mean()

# percentage of students with a passing maths score (50 or greater)
passing_maths = school_data_complete.loc[school_data_complete["maths_score"] >= 50]
percentage_passing_maths = (passing_maths["maths_score"].count() / total_students) * 100

# percentage of students with a passing reading score (50 or greater)
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 50]
percentage_passing_reading = (passing_reading["reading_score"].count() / total_students) * 100

# percentage of students who passed maths and reading (% Overall Passing)
passing_overall = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]
percentage_passing_overall = (passing_overall["student_name"].count() / total_students) * 100

# create a dataframe to hold the above results
local_government_area_summary = pd.DataFrame({"Total Schools": [total_schools],
                                              "Total Students": [total_students],
                                              "Total Budget": [total_budget],
                                              "Average Maths Score": [average_maths_score],
                                              "Average Reading Score": [average_reading_score],
                                              "% Passing Maths": [percentage_passing_maths],
                                              "% Passing Reading": [percentage_passing_reading],
                                              "% Overall Passing": [percentage_passing_overall]})

# give the displayed data cleaner formatting

lga_format_mappings = {
    **common_format_mappings,
    "Total Budget": "${:,.2f}"
}

local_government_area_summary_formatted = format_mappings(local_government_area_summary, lga_format_mappings)

# print the dataframe
local_government_area_summary_formatted
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# school name
school_name = school_data_complete["school_name"].unique()

# school type
school_type = school_data_complete.groupby(["school_name"])["type"].unique().str[0]

# total students
total_students = school_data_complete.groupby(["school_name"])["student_name"].count()

# total school budget
total_school_budget = school_data_complete.groupby(["school_name"])["budget"].unique().str[0]

# per student budget
per_student_budget = total_school_budget / total_students

# average maths score
average_maths_score = school_data_complete.groupby(["school_name"])["maths_score"].mean()

# average reading score
average_reading_score = school_data_complete.groupby(["school_name"])["reading_score"].mean()

# % passing maths
passing_maths = school_data_complete.loc[school_data_complete["maths_score"] >= 50]
percentage_passing_maths = (passing_maths.groupby(["school_name"])["maths_score"].count() / total_students) * 100

# % passing reading
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 50]
percentage_passing_reading = (passing_reading.groupby(["school_name"])["reading_score"].count() / total_students) * 100

# % overall passing
passing_overall = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]
percentage_passing_overall = (passing_overall.groupby(["school_name"])["student_name"].count() / total_students) * 100

# create a dataframe to hold the above results
school_summary = pd.DataFrame({"School Type": school_type,
                               "Total Students": total_students,
                               "Total School Budget": total_school_budget,
                               "Per Student Budget": per_student_budget,
                               "Average Maths Score": average_maths_score,
                               "Average Reading Score": average_reading_score,
                               "% Passing Maths": percentage_passing_maths,
                               "% Passing Reading": percentage_passing_reading,
                               "% Overall Passing": percentage_passing_overall})

# give the displayed data cleaner formatting
school_format_mappings = {
    **common_format_mappings,
    "Total School Budget": "${:,.2f}",
    "Per Student Budget": "${:,.2f}"
}
school_summary_formatted = format_mappings(school_summary, school_format_mappings)

# print the dataframe
school_summary_formatted
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
def sort_and_format(df, ascending=True, top_n=None):
    """Sort by % Overall Passing and format the resulting dataframe."""
    sorted_df = df.sort_values("% Overall Passing", ascending=ascending)
    
    # Formatting the sorted dataframe
    sorted_mappings = {
        "Total School Budget": "${:,.2f}",
        "Per Student Budget": "${:,.2f}",
        **common_format_mappings
    }
    
    sorted_df = format_mappings(sorted_df, sorted_mappings)
    
    if top_n:
        return sorted_df.head(top_n)
    return sorted_df

top_performing_schools_formatted = sort_and_format(school_summary, ascending=False, top_n=5)

# print the five best-performing schools
top_performing_schools_formatted
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_performing_schools_formatted = sort_and_format(school_summary, top_n=5)

# print the five worst-performing schools
bottom_performing_schools_formatted
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
def group_scores_by_year(data, score_column):
    """
    Groups the data by school name and calculates the mean of the given score column 
    for each year (9th to 12th).
    
    Parameters:
        - data: The complete dataset.
        - score_column: The column name to calculate the mean (e.g. 'maths_score' or 'reading_score').

    Returns:
        - A dataframe with years as columns and school names as index.
    """
    grouped_data = {}
    for year in range(9, 13):  # for years 9 to 12
        year_data = data.loc[data["year"] == year]
        grouped_data[f"{year}th"] = year_data.groupby(["school_name"])[score_column].mean()
    
    return pd.DataFrame(grouped_data)


maths_scores_by_year = group_scores_by_year(school_data_complete, 'maths_score')

# give the displayed data cleaner formatting
year_mappings = {
    "9th": "{:.2f}",
    "10th": "{:.2f}",
    "11th": "{:.2f}",
    "12th": "{:.2f}"
}
maths_scores_by_year_formatted = format_mappings(maths_scores_by_year, year_mappings)

# print the dataframe
maths_scores_by_year_formatted
```

* Perform the same operations as above for reading scores

```python
reading_scores_by_year = group_scores_by_year(school_data_complete, 'reading_score')

# give the displayed data cleaner formatting
reading_scores_by_year_formatted = format_mappings(reading_scores_by_year, year_mappings)

# print the dataframe
reading_scores_by_year_formatted
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
agg_dict = {
    'Average Maths Score': 'mean',
    'Average Reading Score': 'mean',
    '% Passing Maths': 'mean',
    '% Passing Reading': 'mean',
    '% Overall Passing': 'mean'
}

def create_binned_column(df, column_to_bin, bins, labels, new_column_name):
    """Bins a column and adds the binned data as a new column."""
    df[new_column_name] = pd.cut(df[column_to_bin], bins, labels=labels)
    return df

def group_and_format(df, group_by_column):
    """Groups data, aggregates metrics, and formats the resultant DataFrame."""
    grouped_data = df.groupby([group_by_column], observed=True).agg(agg_dict)
    return format_mappings(grouped_data, common_format_mappings)

# Now use the functions to streamline the original code
school_summary = create_binned_column(school_summary, "Per Student Budget", [0, 585, 630, 645, 680], ["<$585", "$585-630", "$630-645", "$645-680"], "Spending Ranges (Per Student)")
scores_by_school_spending_formatted = group_and_format(school_summary, "Spending Ranges (Per Student)")

# print the dataframe
scores_by_school_spending_formatted
```
